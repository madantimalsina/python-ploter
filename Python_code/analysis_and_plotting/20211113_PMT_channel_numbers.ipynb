```python
import numpy as np
import matplotlib.pyplot as plt
import uproot

import matplotlib.colors as colors
import matplotlib
# Pick font here
# matplotlib.rcParams['font.serif'] = "Times"
# Then set this font globally
matplotlib.rcParams['font.family'] = "serif"
matplotlib.rcParams['axes.linewidth'] = 2

matplotlib.rcParams['xtick.major.size'] = 7.5
matplotlib.rcParams['xtick.major.width'] = 1.6
matplotlib.rcParams['xtick.minor.size'] = 4.0
matplotlib.rcParams['xtick.minor.width'] = 1.2

matplotlib.rcParams['ytick.major.size'] = 7.5
matplotlib.rcParams['ytick.major.width'] = 1.6
matplotlib.rcParams['ytick.minor.size'] = 4.0
matplotlib.rcParams['ytick.minor.width'] = 1.2
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,10))

pmt_positions = np.loadtxt("/cvmfs/lz.opensciencegrid.org/LZap/ConditionsDataModel/PMTpositions-LZ-21Oct21.txt", skiprows=1)
top_pmts = (pmt_positions[:,3] == 1.53900e+03)
hg_channels = (pmt_positions[:,0] < 1000)
top_hg = top_pmts & hg_channels

pmt_radius = np.sqrt(pmt_positions[:,1]**2 + pmt_positions[:,2]**2)

outer_ring = pmt_radius > 700
outer_top_hg = top_hg & outer_ring

next_outer = (pmt_radius < 700) & (pmt_radius > 625)
next_outer_top_hg = top_hg & next_outer

pmt_labels_hg = ['{:3d}'.format(int(number)) for number in pmt_positions[top_hg][:,0]]

ax.scatter(pmt_positions[top_hg][:,1], pmt_positions[top_hg][:,2], s=500, facecolor='none', color='xkcd:grey', alpha=0.25, linewidth=2)
ax.scatter(pmt_positions[outer_top_hg][:,1], pmt_positions[outer_top_hg][:,2], s=500, facecolor='none', color='xkcd:red', alpha=1, linewidth=2)
ax.scatter(pmt_positions[next_outer_top_hg][:,1], pmt_positions[next_outer_top_hg][:,2], s=500, facecolor='none', color='xkcd:blue', alpha=1, linewidth=2)

for i, label in enumerate(pmt_labels_hg):
    ax.text(pmt_positions[top_pmts][i,1]-25, pmt_positions[top_pmts][i,2]-10, label)

ax.set_xlim([-800, 800])
ax.set_ylim([-800, 800])

#ax.set_title('All single scatters (nSingleScatters RQ)', fontsize=18)
ax.set_xlabel('X [mm]', fontsize=16)
ax.set_ylabel('Y [mm]', fontsize=16)
ax.tick_params(axis='both', which='both', direction='in', labelsize=14)                                   
ax.minorticks_on()

#ax.legend(fontsize=16, loc=(1.05, 0.7))

plt.show()

print('outer ring', pmt_positions[outer_top_hg][:,0])
print('next outer ring', pmt_positions[next_outer_top_hg][:,0])

#fig.savefig("alpaca_plots/XvY_SS_ROI.pdf", bbox_inches='tight')
#fig.savefig("alpaca_plots/xvy.png", bbox_inches='tight')
```
